```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, 
    channel_shift_range=10., horizontal_flip=True)
```

```python
choosen_image = random.choice(os.listdir('/content/drive/MyDrive/Keras/cat_vs_dog/train/cat'))
```

```python
image_path = '/content/drive/MyDrive/Keras/cat_vs_dog/train/cat/'+choosen_image
```

```python
image = np.expand_dims(plt.imread(image_path),0)
```

```python
plt.imshow(image[0])
```

```python
aug_iter = gen.flow(image, save_to_dir='/content/drive/MyDrive/Keras/cat_vs_dog/augmmented_gen_img/', save_prefix='aug-image-', save_format='jpeg')
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotImages(aug_images)
```
